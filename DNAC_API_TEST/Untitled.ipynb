```python
import requests
import json
import datetime
requests.packages.urllib3.disable_warnings()

print ("Current date and time: ")
print (datetime.datetime.now())
print ('Starting DNA Center Hello World - Simple')
print ('Creating Hard Coded Variables')

DNAC_scheme = 'https://'
DNAC_authority = 'sandboxdnac.cisco.com'
DNAC_port = ':443'
DNAC_path_token = '/dna/system/api/v1/auth/token'
DNAC_path = '/dna/intent/api/v1/network-device'
DNAC_user = 'devnetuser'
DNAC_psw = 'Cisco123!'
#WDNAC_user = input("Username? ")
#DNAC_psw = input("Password? ")
```

```python
print ("Current date and time: ")
print (datetime.datetime.now())
print ('Post Token Request')

token_req_url = DNAC_scheme+DNAC_authority+DNAC_path_token
print (token_req_url)
auth = (DNAC_user, DNAC_psw)
print (type(auth))

req = requests.post(token_req_url, auth=auth, verify=False)
print (dir(req))
print ("API Return Code: " + str(req.status_code))
print ('Request URI: ' + token_req_url)
print ("Username: " + DNAC_user)
resp = req.text
print (resp)
token = req.json()["Token"]
print ("Recieved Token:")
print (token)
print ("Length Token:")
print (len(token))
```

```python
import datetime
print ("Current date and time: ")
print (datetime.datetime.now())
print ('Inventory Request - Network Devices')

req_url = DNAC_scheme+DNAC_authority+DNAC_path
print (req_url)
header_data = {'X-auth-token': token}
resp_devices = requests.request('GET', req_url, headers=header_data, verify=False)
print (resp_devices)
resp_devices_json = resp_devices.json()
print('-----------------------------------')

dev_dict = json.loads(resp_devices.text)
print (dev_dict["response"][0].keys())
print ('----------------------------------')
print ("Response (json):")
print (json.dumps(resp_devices_json, indent=2))
```

```python
print ("Current date and time: ")
print (datetime.datetime.now())
print ('Inventory Request - Filtering output')

for device in resp_devices_json['response']:
    if device['type'] != None:
        print ('---')
        print ('Hostname: ' + device['hostname'])
        print ('Type: ' + device['type'])
        print ('IP: ' + device['managementIpAddress'])
        print ('MAC: ' + device['macAddress'])
        print ('Software Type: ' + device['softwareType'])
        print ('Software Version: ' + device['softwareVersion'])
        print ('UpTime: ' + device['upTime'])
```

```python
print ("Current date and time: ")
print (datetime.datetime.now())
print ('Inventory Request - Filtering output')

reachable = 0
count = 0

for device in resp_devices_json['response']:
    
    if device['type'] != None:
        if device['reachabilityStatus'] == "Reachable":
            reachable += 1
        count += 1
if count == reachable:
    print ("all devices are reachable")
else: print ("not all devices are reachable")
```

```python
import csv

# now we will open a file for writing
resp_devices_csv = open('resp_devices.csv', 'w')
 
# create the csv writer object
csv_writer = csv.writer(resp_devices_csv)
 
# Counter variable used for writing 
# headers to the CSV file
count = 0
 
for device in dev_dict["response"]:
    if count == 0:
 
        # Writing headers of CSV file
        header = device.keys()
        csv_writer.writerow(header)
        count += 1
 
    # Writing data of CSV file
    csv_writer.writerow(device.values())
 
resp_devices_csv.close()
```

```python
import datetime
import csv

# Create a CSV file with current timestamp in the name
current_time = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
csv_filename = f"inventory_report_{current_time}.csv"

# Open the CSV file for writing
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Write the header row
    writer.writerow([
        'Hostname', 'Type', 'IP', 'MAC', 'Software Type', 'Software Version', 'UpTime'
    ])

    # Write the device data to the CSV
    for device in resp_devices_json['response']:
        if device['type'] is not None:
            writer.writerow([
                str(device.get('hostname', 'N/A')),
                str(device.get('type', 'N/A')),
                str(device.get('managementIpAddress', 'N/A')),
                str(device.get('macAddress', 'N/A')),
                str(device.get('softwareType', 'N/A')),
                str(device.get('softwareVersion', 'N/A')),
                str(device.get('upTime', 'N/A'))
            ])

print(f"CSV file '{csv_filename}' has been created with the device data.")
```

```python
from flask import Flask, render_template
import datetime

app = Flask(__name__)

@app.route('/')
def index():
    # Filter device data (only devices with a non-null 'type')
    devices = [
        {
            'hostname': device.get('hostname', 'N/A'),
            'type': device.get('type', 'N/A'),
            'ip': device.get('managementIpAddress', 'N/A'),
            'mac': device.get('macAddress', 'N/A'),
            'software_type': device.get('softwareType', 'N/A'),
            'software_version': device.get('softwareVersion', 'N/A'),
            'uptime': device.get('upTime', 'N/A')
        }
        for device in resp_devices_json['response'] if device['type'] is not None
    ]

    # Render the HTML template and pass the device data to it
    return render_template('index.html', devices=devices)


if __name__ == '__main__':
    app.run(debug=True)
```
